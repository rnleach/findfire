```python
import satfire as sf
import matplotlib.pyplot as plt
from datetime import datetime as dt, date, timedelta

plt.rcParams['figure.figsize'] = [20, 10]

db = sf.SatfireDatabases()
```

```python
g16 = db.total_fire_power_time_series(8484265)
g17a = db.total_fire_power_time_series(2208150)
g17b = db.total_fire_power_time_series(8512746)
```

```python
f, ax = plt.subplots()

ax.plot(g16['scan start'], g16['total power'], label='G16')
ax.plot(g17a['scan start'], g17a['total power'], label='G17A')
ax.plot(g17b['scan start'], g17b['total power'], label='G17B')

ax.set_xlim((dt(2022,4,6), dt(2022,6,12)))

ax.legend()
```
